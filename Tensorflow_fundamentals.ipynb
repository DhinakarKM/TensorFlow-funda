```python
abc = tf.constant(np.random.randint(0,1000,size=100))
abc
```

```python
np.min(abc)
tf.reduce_min(abc)
tf.reduce_max(abc)
tf.reduce_mean(abc)
tf.reduce_sum(abc)
tf.math.reduce_variance(tf.cast(abc, dtype=tf.float32)) # Cast to float32 for variance calculation
tf.math.reduce_std(tf.cast(abc, dtype=tf.float32)) # Cast to float32 for standard deviation calculation
```

```python
import tensorflow_probability as tfp
tfp.stats.variance(abc)
tfp.stats.stddev(tf.cast(abc,dtype=tf.float32))
```
